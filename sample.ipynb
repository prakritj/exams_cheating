```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
mean = 72
stdev = 10
n = 1000
```

```python
norm_exam = stats.skewnorm.rvs(-1,size = n, loc = mean, scale = stdev)
sns.histplot(x=norm_exam, bins = 50, kde = True, stat ='density')
```

I need axis values to be b/w 0 and 100 - but you can see here that I have some values over 100. How do I get rid of them? Two options:
- can replace all values > 100 with just 100
- can replace all values > 100 with a random value from the same normally distributed dist which is < 100

```python
sns.ecdfplot(norm_exam)
```
